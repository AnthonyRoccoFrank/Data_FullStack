```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import datetime as dt
```

Vaccines Transform

```python
#Reading in vaccines2018 to Dataframe
vaccines = pd.read_csv("ExtractCSV/vaccines/vaccines2018.csv")
vaccines
```

```python
#Transforming DF to usable data
#Number of doses given
vaccines["Total Doses Distributed"] = vaccines["Total Doses Distributed"].str.replace(" M","")
vaccines["Total Doses Distributed"] = vaccines["Total Doses Distributed"].str.replace("≈ ","")
vaccines["Total Doses Distributed"] = vaccines["Total Doses Distributed"].str.replace("≈","")
vaccines["Total Doses Distributed"] = vaccines["Total Doses Distributed"].astype(float)
#Number changing from string to date
vaccines["Week"] = pd.to_datetime(vaccines["Week"])
vaccines['Year'] = vaccines['Week'].dt.year
vaccines['Month'] = vaccines['Week'].dt.month
vaccines['Day'] = vaccines['Week'].dt.day
vaccines
```

```python
vaccines_11to17 = pd.DataFrame()

for k in range(7):
    try:
        addayear = pd.read_csv("ExtractCSV/vaccines/vaccines201" + str(k+1) + ".csv")
        vaccines_11to17 = pd.concat([vaccines_11to17, addayear], ignore_index=True)

    except Exception as e:
        print(f'An error occurred: {e} in {k}')

vaccines_11to17
```

```python
approximate = "≈ = approximate"
vaccines_11to17.drop(vaccines_11to17.loc[(vaccines_11to17["Week"]== approximate),:].index,inplace=True)
```

```python
#Transforming DF to usable data
#Number of doses given
vaccines_11to17["Total Doses Distributed"] = vaccines_11to17["Total Doses Distributed"].str.replace(" million doses","")
vaccines_11to17["Total Doses Distributed"] = vaccines_11to17["Total Doses Distributed"].str.replace("≈ ","")
vaccines_11to17["Total Doses Distributed"] = vaccines_11to17["Total Doses Distributed"].astype(float)
#Number changing from string to date
vaccines_11to17["Week"] = pd.to_datetime(vaccines_11to17["Week"])
vaccines_11to17['Year'] = vaccines_11to17['Week'].dt.year
vaccines_11to17['Month'] = vaccines_11to17['Week'].dt.month
vaccines_11to17['Day'] = vaccines_11to17['Week'].dt.day
vaccines_11to17
```

```python
vaccines_20to21 = pd.DataFrame()

for k in range(0,1):
    try:
        addayear = pd.read_csv("ExtractCSV/vaccines/vaccines202" + str(k) + ".csv")
        vaccines_20to21 = pd.concat([vaccines_20to21, addayear], ignore_index=True)

    except Exception as e:
        print(f'An error occurred: {e} in {k}')

vaccines_20to21
```

```python
#Transforming DF to usable data
#Number of doses given
vaccines_20to21["Total Doses Distributed"] = vaccines_20to21["Total Doses Distributed"].str.replace(" M","")
vaccines_20to21["Total Doses Distributed"] = vaccines_20to21["Total Doses Distributed"].str.replace("≈ ","")
vaccines_20to21["Total Doses Distributed"] = vaccines_20to21["Total Doses Distributed"].str.replace("*","")
vaccines_20to21["Total Doses Distributed"] = vaccines_20to21["Total Doses Distributed"].astype(float)
```

```python
vaccines_20to21['Year'] = vaccines_20to21['Week'].astype(str)
```

```python
#Number changing from string to date
vaccines_20to21["Week"] = pd.to_datetime(vaccines_20to21["Week"])
vaccines_20to21['Year'] = vaccines_20to21['Week'].dt.year
vaccines_20to21['Month'] = vaccines_20to21['Week'].dt.month
vaccines_20to21['Day'] = vaccines_20to21['Week'].dt.day
vaccines_20to21
```

```python
vaccines_11to21 = pd.concat([vaccines_11to17,vaccines,vaccines_20to21], ignore_index=True)
vaccines_11to21 = vaccines_11to21.sort_values('Week').reset_index(drop=True)
vaccines_11to21.to_csv("TransformCSV/vaccines_11to21.csv")
vaccines_11to21
```
